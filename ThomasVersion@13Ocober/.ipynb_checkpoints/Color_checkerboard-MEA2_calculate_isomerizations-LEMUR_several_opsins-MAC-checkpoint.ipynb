```python
def load_obj(name ):
    if name[-4:]=='.pkl':
        name = name[:-4]
    #~ try:
        #~ return pk5.dumps(name+'pkl', protocol=5)
    #~ except:
    with open( name + '.pkl', 'rb') as f:
        return pickle.load(f)
def save_obj(obj, name ):
    if name[-4:]=='.pkl':
        name = name[:-4]
    with open( name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
from scipy import interpolate
start_nm    = 300
end_nm      = 700
wavelengths = np.arange(start_nm, end_nm, 1)

Lambdas = np.array(range(300,701))
wavlengthmin=300   #in nm
wavlengthmax=700    #in nm
Scones=1
Mela=1
Rods=1
Mcones=1
ReachR=1
wl=np.linspace(wavlengthmin,wavlengthmax,wavlengthmax-wavlengthmin+1)
xnew = np.linspace(350,700,701)
#Extract Spectrum Data from txt and increase the range of the data to a specific set of wavelengths (wl)
#They need to have one value for each wavelength on a specific range of wavelength

def ExtractDatafromtxt(dir,nwl,nAmp,wl):
    Amp=[]
    data=[]
    with open(dir,'r') as file:
        for i in file:
            data.append(i.split('\t'))
            
        wldata,Ampdata=[],[]
        for dat in data:
            wldata.append(dat[nwl])
            Ampdata.append(dat[nAmp])
        
        i_wldat=0
        for i_wl in range(len(wl)):
            if i_wldat<len(wldata)-1:
                if wl[i_wl]==float(wldata[i_wldat]):
                    Amp.append(float(Ampdata[i_wldat]))
                    i_wldat+=1
                elif wl[i_wl]<float(wldata[i_wldat]):
                    Amp.append(0)
                elif wl[i_wl]>float(wldata[i_wldat]):
                    i_wldat+=int(wl[i_wl]-float(wldata[i_wldat]))
                    Amp.append(float(Ampdata[i_wldat]))
                    i_wldat+=1
            else:
                Amp.append(0)
    return(Amp)                    

#Extraction and plotting of the opsins spectrum
dir='../Spectre_photorecepteur/'

print('Opsins Spectrum')

AmpScones=ExtractDatafromtxt(dir + "mouse_cone_opsins.txt",2,0,wl)
AmpMcones=ExtractDatafromtxt(dir + "mouse_cone_opsins.txt",2,1,wl)
AmpRods=ExtractDatafromtxt(dir + "rods.txt",0,1,wl)
AmpMela=ExtractDatafromtxt(dir + "melanopsins.txt",1,0,wl)
AmpReach=ExtractDatafromtxt(dir + "ReachR.txt",0,1,wl)

AmpPhoto=[]
fig=plt.figure(figsize=(8,4))
if Scones==1:
    plt.plot(wl, AmpScones, 'b',lw=3,label='Scones')
    f = interpolate.interp1d(wl,AmpScones)
    AmpScones = f(xnew)/np.max(f(xnew))
    AmpPhoto.append(AmpScones)
if Mela==1:
    plt.plot(wl, AmpMela,'m',lw=2,label='Mela')
    f = interpolate.interp1d(wl,AmpMela)
    AmpMela = f(xnew)/np.max(f(xnew))
    AmpPhoto.append(AmpMela)
if Rods==1:
    plt.plot(wl, AmpRods, 'k',lw=2,label='Rods')
    f = interpolate.interp1d(wl,AmpRods)
    AmpRods = f(xnew)/np.max(f(xnew))
    AmpPhoto.append(AmpRods)
if Mcones==1:
    plt.plot(wl, AmpMcones, 'g',lw=3,label='Mcones')
    f = interpolate.interp1d(wl,AmpMcones)
    AmpMcones = f(xnew)/np.max(f(xnew))
    AmpPhoto.append(AmpMcones)
if ReachR==1:
    plt.plot(wl, AmpReach, 'r',lw=3)
    f = interpolate.interp1d(wl,AmpReach)
    AmpReach = f(xnew)/np.max(f(xnew))
    AmpPhoto.append(AmpReach)

legend()

colors = ['b','k','r']

OpsinsLemur = []
for l in np.arange(3):
    f = interpolate.interp1d(Lambdas,SpectrumAll[l])
    Amps = f(xnew)/np.max(f(xnew))
    OpsinsLemur.append(Amps)
    
    plot(Lambdas,SpectrumAll[l],'--',color=colors[l])
    
```

```python
# Franke code, from Euler lab
import spectrum as spc

from spectrum import *
    
def setPlotStyle():
    # Presettings for figures
    #
    mpl.rcParams['figure.figsize'] = [10, 5]
    mpl.rc('font', size=10)          
    mpl.rc('axes', titlesize=12)    
    mpl.rc('axes', labelsize=12)    
    mpl.rc('xtick', labelsize=12)   
    mpl.rc('ytick', labelsize=12)   
    mpl.rc('legend', fontsize=12)   
    mpl.rc('figure', titlesize=12) 

start_nm    = 300
end_nm      = 700
wavelengths = np.arange(start_nm, end_nm, 1)

def eq8(x):
    # Define constants
    #
    a = -188862.970810906644
    b = 90228.966712600282
    c = -2483.531554344362
    d = -6675.007923501414
    e = 1813.525992411163
    f = -215.177888526334
    g = 12.487558618387
    h = -0.289541500599
    return 10**(a + b*x**2 + c*x**4 + d*x**6 + e*x**8 + f*x**10 + g*x**12 + h*x**14)

# Calculate x 
#
x = np.log10(wavelengths)

# Generate opsin spectrum Atemplate
#
temp = []
for i in range(end_nm -start_nm):
    value = eq8(x[i])
    temp.append(value)
spect_template = np.array(temp)    
    
# Find wavelength at peak
#
peak    = max(temp)
peak_nm = temp.index(peak) +start_nm
spect_template = np.array(temp)    
  
# Plot template
#
setPlotStyle()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
col = wavelength_to_rgb(peak_nm, darker=0.75)
ax1.semilogy(wavelengths, spect_template, color=col, label="opsin template")
ax1.set_xlabel("wavelength [nm]")        
ax1.set_ylabel("norm. sensitivity [log]")      
ax1.set_ylim((0.005, 1.0))
ax1.set_xlim((300, 700))
ax2.plot(wavelengths, spect_template, color=col, label="opsin template")
ax2.set_ylim((0.005, 1.0))
ax2.set_ylabel("rel. sensitivity")      
ax2.legend(bbox_to_anchor=(1.1, 1), loc="upper left")
ax2.grid(False)   
```

```python
# Olivier calculations

# 501 nm for rhodopsin. As a reminder, it is 409 nm for S opsin and 558 nm for L opsin

SpectrumAll=[]
i=0
for l in [409, 501, 558]:#:[409,501,558]:

    Lambdas = np.array(range(300,701))
    LambdaMax = l      #558.

    A= 69.7
    a=0.8795 + 0.0459*np.exp( (-1*(LambdaMax - 300)**2 )/11940 )
    B=28
    b=0.922
    C=-14.9
    c=1.104
    D=0.674
    Asupp=0.26
    Lsupp=189+0.315*LambdaMax
    bsupp = -40.5+0.195*LambdaMax


    x = LambdaMax / Lambdas
    MainSp = 1 / ( np.exp(A*(a-x)) + np.exp(B*(b-x)) + np.exp(C*(c-x)) + D) 
    LeftSupp = Asupp*np.exp( -1*((Lambdas - Lsupp)/bsupp)**2 )   
    LeftSupp = np.where(Lambdas<LambdaMax, LeftSupp,0*LeftSupp)
    Spectrum = MainSp + LeftSupp

    SpectrumAll.append(Spectrum)
    plot(Lambdas,Spectrum,color=colors[i])
    i+=1
    
# plt.plot(wavelengths, mouse_spect[0],'--' ,color=col, label=mouse_opsins[iP])
# plt.plot(wavelengths, mouse_spect[1],'--' ,color=col, label=mouse_opsins[iP])
# plt.plot(wavelengths, mouse_spect[2],'--' ,color=col, label=mouse_opsins[iP])

fsave = './PhotoReceptorData'
PhData = load_obj(fsave)

Opsins=[]
Opsins.append(PhData['Scones'])
Opsins.append(PhData['Mela'])
Opsins.append(PhData['Rods'])
Opsins.append(PhData['Mcones'])
Opsins.append(PhData['RedOpsin'])

x = PhData['x_axis']

for o in Opsins[2:3]:
    plot(x,o,'--')
```

# Calculate Isomerization for one source for all opsins

```python
#Illuminance Factor Product
"""""
Main function that gets: 
1 - the light amplitude not normalized in µW/cm²
2 - the spectra of the target opsin normalized to 1 at the peak (a.u.)
3 - the wavelength step (nm)

Factors have units of:
ac = µm²
hc = J.m
10**23 puts all in µW - µm² 

Output is in isomerizations / s

To get the result in photons/cm^2/s we have to multiply by 10^8 and divide by ac (x5 or x2 for rods)
"""""
#hyperparameters
h=6.63*10**(-34) #en J.s
c=299792458      #en m/s
ac=[0.2,0.5,0.2]           # en µm²
# ac_cones=0.2
# ac_rods=0.5
# ac_mela=0.2
# ac_reach=0.2

def IlluminanceFactor(Amp1,AmpPhoto,wl):   
    # Amplitude 1 is the LED source normalized to have the integral equal to the total power. 
    # Amp photo is the Spectral curve of the photoreceptor (all 4 together)
    # wl is the steps in lambda for the curves
    # Gives back the result of each photoreceptor
    
    IlluminanceFactor=[]
    for photo in range(len(AmpPhoto)):
        IlFa=0
        for I_wl in range(len(Amp1)):
            # Results in photoisomerisation/s /µm²
            IlFa+= Amp1[I_wl] * AmpPhoto[photo][I_wl] * wl[I_wl] * diff(wl)[10]  * ac[photo]/(h*c) *10**(-23) 
                    #µW/cm²   x   a.u.(peak at 1)    x nm (lambda) x lambda step x ph.µm² x 1/(Jm)  x convert to same units
        IlluminanceFactor.append(IlFa) 
    return np.array(IlluminanceFactor) #

def Write_fancy(result,F):
    o_names = ['Scones    ','Melanopsin','Rhodopsin ','Mcones    ','Red_opsin ']
    for i,r in enumerate(result):
        if F=='scientific':
            print(o_names[i],' :',"{:.2e}".format(r))
        else:
             print(o_names[i],' :',"{:10.0f}".format(r))

                
def Write_fancy_lemur(result,F):
    o_names = ['Scones    ','Rhodopsin ','Mcones    ']
    for i,r in enumerate(result):
        if F=='scientific':
            print(o_names[i],' :',"{:.2e}".format(r))
        else:
             print(o_names[i],' :',"{:10.0f}".format(r))

    
```

```python
# Load LED data

file_cal = rootf+'/matias/Dropbox/Drop-IdV/PROJECTS/COLORS/EXPS_AND_CALIBRATIONS/Calibration_MEA2/cal_with_MEA2'
caldata =load_obj(file_cal)

print(caldata.keys())
lx = caldata['xnew']
lx[0] = 300

color_order = caldata['orderS']
spec_curves = np.array(caldata['LedSpectra'])
spec_curves[:,0]=0


print(caldata['orderV'])
reorder = [1,2,3,4,5]
Vcurves = np.array(caldata['Vcurves'])[:,reorder]

#---------------------------------------------
# New interpolated V calibration curves

Vnew = np.linspace(0,5,5001)
newVcurves = np.zeros([5,5001])
for i in np.arange(5):
    vals = (np. isnan(Vcurves[:,i]))==False
    f = interpolate.interp1d(caldata['V'][vals],Vcurves[:,i][vals],fill_value="extrapolate")
    newVcurves[i] = f(Vnew) / correction_f  # use interpolation function returned by `interp1d`
    newVcurves[i] = np.where(newVcurves[i]<0,0,newVcurves[i]) # take negative extrapolations out

# newVcurves[4][0:34] = 0  # no negative values


Rmax = np.nanmax(newVcurves[0])
Ymax = np.nanmax(newVcurves[1])
Gmax = np.nanmax(newVcurves[2])
Bmax = np.nanmax(newVcurves[3])
Vmax = np.nanmax(newVcurves[4])

# Function that returns a voltage given a certain power value in percentage of max possible value  
def find_V(value,color):
    if color =='r':
            array = newVcurves[0]
            vmax=Rmax
    elif color== 'y':
            array = newVcurves[1]
            vmax=Ymax
    elif color== 'g':
            array = newVcurves[2]
            vmax=Gmax
    elif color== 'b':
            array = newVcurves[3]
            vmax=Bmax
    elif color== 'v':
            array = newVcurves[4]
            vmax=Vmax
    
    array = np.asarray(array)
    idx = (np.abs(array - value*vmax)).argmin()
    return Vnew[idx]
#==============
colors = ['violet','b','g','y','r']

for i in np.arange(0,5):
    plot(caldata['V'],Vcurves[:,i]/nanmax(Vcurves[:,i]),'-o',color=colors[4-i])

figure()
plot(newVcurves[0])
figure()
for i in np.arange(1,5):
    plot(newVcurves[i],colors[4-i])
```
